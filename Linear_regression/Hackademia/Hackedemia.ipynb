```python
X=np.array(data['Hours'])
Y=np.array(data['Scores'])
plt.scatter(X,Y)
plt.xlabel('Hours')
plt.ylabel('Grades')
```

```python
def compute(X,Y):
    n=len(X)
    X_mean=np.mean(X)
    Y_mean=np.mean(Y)
    num=0
    den=0
    for i in range(0,len(X)):
       num+=((X[i]-X_mean)*(Y[i]-Y_mean))
       den+=(X[i]-X_mean)**2
    slope=num/den   
    intercept=Y_mean-slope*X_mean   
    return slope,intercept

slope,intercept=compute(X,Y)
print(slope,intercept)
```

```python
Y_pred=[slope*i +intercept for i in X]
```

```python
Y_mean=np.mean(Y)
nume=0
deno=0
for i in range(0,len(Y)):
    nume+=(Y[i]-Y_pred[i])**2
    deno+=(Y[i]-Y_mean)**2
r2=1-nume/deno    
print(r2)
```
